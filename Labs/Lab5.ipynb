### Кривая Коха

```python
import turtle
import time

n = int(input('Введите глубину фрактала: '))
def koch_curve(ln, n):
    if n == 0:
        t.fd(ln)
        return
    a = ln / 3
    koch_curve(a, n - 1)
    t.left(60)
    koch_curve(a, n - 1)
    t.right(120)
    koch_curve(a, n - 1)
    t.left(60)
    koch_curve(a, n - 1)


t = turtle.Turtle()
t.ht()

start_time = time.time()
koch_curve(200, n)
print(" %s " % (time.time() - start_time))

turtle.done()
```

### Кривая Гильберта

```python
n = int(input('Введите глубину фрактала: '))
def hilbert_curve(n, a, ln):
    if n == 0:
        return
    
    t.right(a)
    hilbert_curve(n-1, -a, ln)
  
    t.fd(ln)
    t.left(a)
    hilbert_curve(n-1, a, ln)
  
    t.fd(ln)
    hilbert_curve(n-1, a, ln)
  
    t.left(a)
    t.fd(ln)
    hilbert_curve(n-1, -a, ln)
    t.right(a)
    
t = turtle.Turtle()
t.ht()
    
start_time = time.time()
hilbert_curve(n, 90, 10)
print(" %s " % (time.time() - start_time))

turtle.done()
```

### Кривая Серпинского

```python
n = int(input('Введите глубину фрактала: '))
def half_sierpinski_curve(ln, n):
    if n == 0:
        return
    half_sierpinski_curve(ln, n-1)
    t.left(45)
    t.fd(ln*(2**(0.5)))
    t.left(45)
    half_sierpinski_curve(ln, n-1)
    t.right(90)
    t.fd(ln)
    t.right(90)
    half_sierpinski_curve(ln, n-1)
    t.left(45)
    t.fd(ln*(2**(0.5)))
    t.left(45)
    half_sierpinski_curve(ln, n-1)
    
def sierpinski_curve(ln, n):
    half_sierpinski_curve(ln, n)
    t.right(90)
    t.fd(ln)
    t.right(90)
    half_sierpinski_curve(ln, n)
    t.right(90)
    t.fd(ln)
    t.right(90)
    
    
t = turtle.Turtle()
t.ht()

start_time = time.time()
sierpinski_curve(10, n)
print(" %s " % (time.time() - start_time))

turtle.done()
```

### Салфетка Серпинского

```python
n = int(input('Введите глубину фрактала: '))
def sierpinski(ln, n):
    if n == 0:
        t.stamp()
    else:
        for i in range(0, 3):
            t.fd(ln)
            sierpinski(ln/2, n-1)
            t.bk(ln)
            t.left(120)
            
            
t = turtle.Turtle()

t.right(30)
t.pu()
t.ht()

t.shape('triangle')
    
start_time = time.time()
sierpinski(100, n)
print(" %s " % (time.time() - start_time))

turtle.done()
```

### Ковер Серпинского

```python
n = int(input('Введите глубину фрактала: '))
def sierpinski_carpet(n, m):
    if n == 0: 
        t.color('black')
        t.begin_fill()
        for i in range(4):
            t.forward(m)
            t.left(90)
        t.end_fill()
    else:
        for i in range(4):
            sierpinski_carpet(n-1, m/3)
            t.forward(m/3)
 
            sierpinski_carpet(n-1, m/3)
            t.forward(m/3)
 
            t.forward(m/3)
            t.left(90)

        
t = turtle.Turtle()
t.ht()

start_time = time.time()
sierpinski_carpet(n, 100)
print("%s " % (time.time() - start_time))

turtle.done()
```

### Дерево Пифагора 

```python
n = int(input('Введите глубину фрактала: '))
def pythagoras_tree(ln, n): 
    if n <= 0 :  
        t.fd(ln) 
    else:
        t.right(135)
        t.fd((ln*ln/2)**(0.5))
        t.left(90)
        pythagoras_tree(((ln*ln/2)**(0.5)), n-1)
        t.left(90)
        t.fd((ln*ln/2)**(0.5))
        t.left(90)
        t.fd((ln*ln/2)**(0.5))
        t.right(135)
        t.fd(ln)
        t.right(135)
        t.fd((ln*ln/2)**(0.5))
        t.left(90)
        t.fd((ln*ln/2)**(0.5))
        t.left(90)
        pythagoras_tree(((ln*ln/2)**(0.5)), n-1)
        t.left(90)
        t.fd((ln*ln/2)**(0.5))
        t.right(135)
        
def carre(ln, n):
    t.fd(ln)
    t.left(90)
    t.fd(ln)
    t.left(90)
    pythagoras_tree(ln, n)
    t.left(90)
    t.fd(ln)
       
t = turtle.Turtle()
t.ht()

start_time = time.time()
carre(100, n)
print(" %s " % (time.time() - start_time))

turtle.done()
```
