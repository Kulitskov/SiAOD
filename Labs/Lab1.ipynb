#### Выборка

```python
import math
import time

def sel_sort(mas):
    for i in range(len(mas) - 1):
        m = i
        j = i + 1
        while j < len(mas):
            if mas[j] < mas[m]:
                m = j
            j+=1
        mas[i], mas[m] = mas[m], mas[i]
        
mas_rand = [random.randint(1, 99) for i in range(10)]
 
print(mas_rand)
time_start = time.perf_counter()
sel_sort(mas_rand)
time_end = time.perf_counter()
print(mas_rand, f"\nСортировка заняла {time_end - time_start:0.10f} секунд")
```

#### Вставка

```python
def insertion_sort(mas):
    for i in range(1, len(mas)):
        temp = mas[i]
        j = i - 1
        while (j >= 0 and temp < mas[j]):
            mas[j+1] = mas[j]
            j-=1
        mas[j+1] = temp
        
mas_rand = [random.randint(1, 99) for i in range(10)]

print(mas_rand)
time_start = time.perf_counter()
insertion_sort(mas_rand)
time_end = time.perf_counter()
print(mas_rand, f"\nСортировка заняла {time_end - time_start:0.10f} секунд")
```

#### Обмен

```python
def bubble(mas):
    for i in range(len(mas)-1):
        for j in range(len(mas)-i-1):
            if mas[j] > mas[j+1]:
                mas[j], mas[j+1] = mas[j+1], mas[j]
                
mas_rand = [random.randint(1, 99) for i in range(10)]

print(mas_rand)
time_start = time.perf_counter()
bubble(mas_rand)
time_end = time.perf_counter()
print(mas_rand, f"\nСортировка заняла {time_end - time_start:0.10f} секунд")
```

#### Шелл

```python
def shellSort(mas):
    n = len(mas)
    k = int(math.log2(n))
    interval = 2**k -1
    while interval > 0:
        for i in range(interval, n):
            temp = mas[i]
            j = i
            while (j >= interval) and (mas[j - interval] > temp):
                mas[j] = mas[j - interval]
                j -= interval
            mas[j] = temp
        k -= 1
        interval = 2**k -1
    return mas

mas_rand = [random.randint(1, 99) for i in range(10)]

print(mas_rand)
time_start = time.perf_counter()
shellSort(mas_rand)
time_end = time.perf_counter()
print(mas_rand, f"\nСортировка заняла {time_end - time_start:0.10f} секунд")
```

#### Быстрая сортировка

```python
def quicksort(array):
    if(len(array)<2):
        return array
    else:
        pivot = array[0]
        less = [i for i in array[1:] if i <= pivot]
        grater = [i for i in array[1:] if i > pivot]
        
        return quicksort(less) + [pivot] + quicksort(grater)

mas_rand = [random.randint(1, 99) for i in range(10)]

print(mas_rand)
time_start = time.perf_counter()
mas_rand = quicksort(mas_rand)
time_end = time.perf_counter()
print(mas_rand, f"\nСортировка заняла {time_end - time_start:0.10f} секунд")
```

#### Пирамидальная

```python
def heapsort(alist):
    build_max_heap(alist)
    for i in range(len(alist) - 1, 0, -1):
        alist[0], alist[i] = alist[i], alist[0]
        max_heapify(alist, index=0, size=i)
 
def parent(i):
    return (i - 1)//2
 
def left(i):
    return 2*i + 1
 
def right(i):
    return 2*i + 2
 
def build_max_heap(alist):
    length = len(alist)
    start = parent(length - 1)
    while start >= 0:
        max_heapify(alist, index=start, size=length)
        start -= 1
 
def max_heapify(alist, index, size):
    l = left(index)
    r = right(index)
    if (l < size and alist[l] > alist[index]):
        largest = l
    else:
        largest = index
    if (r < size and alist[r] > alist[largest]):
        largest = r
    if (largest != index):
        alist[largest], alist[index] = alist[index], alist[largest]
        max_heapify(alist, largest, size)
 
 
mas_rand = [random.randint(1, 99) for i in range(10)]

print(mas_rand)
time_start = time.perf_counter()
heapsort(mas_rand)
time_end = time.perf_counter()
print(mas_rand , f"\nСортировка заняла {time_end - time_start:0.10f} секунд")
```
