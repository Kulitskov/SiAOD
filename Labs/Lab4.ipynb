● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>

```python
class Stack():
    def __init__(self):
        self.stack = []
    def append(self, item):
        self.stack.append(item)
    def pop(self):
        return self.stack.pop()
    def isnotempty(self):
        if len(self.stack) == 0:
            return 0
        else:
            return 1
```

```python
class Deque():
    def __init__(self):
        self.deque = []
    def append(self, item, nap = 0):
        self.deque.insert(nap, item)
    def pop(self, nap = -1):
        return self.deque.pop()
    def isnotempty(self):
        if len(self.deque) == 0:
            return 0
        else:
            return 1
```

```python
class stack():
    def __init__(self, data = []):
        self.stack = data
    def is_empty(self):
        return False if len(self.stack) else True
    def append(self, item):
        self.stack.append(item)
    def pop(self):
        return self.stack.pop()
    def __str__(self):
        result = ""
        for el in self.stack:
            result += str(el) + " "
        return "[" + result[:-1] + "]"

class deque():
    def __init__(self, data = []):
        self.deque = data
    def is_empty(self):
        return False if len(self.deque) else True
    def append_l(self, item):
        self.deque.insert(0, item)
    def append_r(self, item):
        self.deque.append(item)
    def pop_l(self):
        return self.deque.pop(0)
    def pop_r(self):
        return self.deque.pop()
    def __str__(self):
        result = ""
        for el in self.deque:
            result += str(el) + " "
        return "[" + result[:-1] + "]"
```

### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
def func(data):
    deq1 = deque([])
    deq2 = deque([])
    if len(data) == 0:
        return []
    deq1.append_l(data[0])
    for i in range(1, len(data)):
        temp1 = deq1.pop_l()
        if data[i] <= temp1:
            deq1.append_l(temp1)
            deq1.append_l(data[i])
            continue
        deq1.append_l(temp1)
        temp2 = deq1.pop_r()
        if data[i] >= temp2:
            deq1.append_r(temp2)
            deq1.append_r(data[i])
            continue
        deq1.append_r(temp2)
        temp = deq1.pop_l()
        while data[i] >= temp:
            deq2.append_r(temp)
            temp = deq1.pop_l()
        deq1.append_l(temp)
        deq1.append_l(data[i])
        while deq2.is_empty() is False:
            deq1.append_l(deq2.pop_r())
    return deq1.deque

try:
    with open("F1.txt", "r", encoding = "utf-8") as file:
        temp = []
        for line in file:
            temp.append(line)
        temp[-1] += "\n"
        result = func(temp)
except FileNotFoundError:
    print("Файл не найден")
try:
    with open("NewF1.txt", "w", encoding = "utf-8") as file:
        file.writelines(result)
except:
    print("Что-то поломалось")
```


### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
def num4():
    f = open('F4.txt', 'r')
    b = False
    stack = Stack()
    for i in f:
        b = True
        x = i
    if b:
        for i in range(len(x)):
            if x[i] == "[":
                stack.append(x[i])
            elif x[i] == "]":
                if stack.isnotempty():
                    stack.pop()
                else:
                    return "No"
        if stack.isnotempty():
            return "No"
        else:
            return "Yes"
    else:
        return "Yes"

print(num4())
```

### Задание №5

```python
def num5():
    f = open('F5.txt', 'r')
    b = False
    deque = Deque()
    for i in f:
        b = True
        x = i
    if b:
        for i in range(len(x)):
            if x[i] == "[":
                deque.append(x[i])
            elif x[i] == "]":
                if deque.isnotempty():
                    deque.pop()
                else:
                    return "No"
        if deque.isnotempty():
            return "No"
        else:
            return "Yes"
    else:
        return "Yes"

print(num5())
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
stackC = Stack()
stackB = Stack()
stackS = Stack()
c = [str(i) for i in range(10)]
f = open('F6.txt', 'r')
for i in f:
    x = i.split()
    x.reverse()
for i in range(len(x)):
    if len(x[i]) != 1 and x[i][0] == "-" and x[i][1] in c:
        stackC.append(x[i])
    elif len(x[i]) != 1 and x[i][0] == "-" and x[i][1] not in c:
        stackS.append(x[i])
    elif x[i][0] in c:
        stackC.append(x[i])
        
    elif ord(x[i][0]) >= 65 and ord(x[i][0]) <= 90:
        stackB.append(x[i])
    elif ord(x[i][0]) >= 97 and ord(x[i][0]) <= 122:
        stackB.append(x[i])
    elif ord(x[i][0]) >= 192 and ord(x[i][0]) <= 255:
        stackB.append(x[i])
    
    else:
        stackS.append(x[i])
while stackC.isnotempty():
    print(stackC.pop(), end = " ")
while stackB.isnotempty():
    print(stackB.pop(), end = " ")
while stackS.isnotempty():
    print(stackS.pop(), end = " ")
f.close()

#  or (chr(x[i][0]) >= 97 and chr(x[i][0] <= 122) or (chr(x[i][0]) >= 192 and chr(x[i][0] <= 255)
```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.

```python
dequeP = Deque()
dequeM = Deque()
f = open('F7.txt', 'r')
nf = open('NF7.txt', 'w')
for i in f:
    s = i.split()
for i in range(len(s)):
    if int(s[i]) > 0:
        dequeP.append(s[i])
    elif int(s[i]) < 0:
        dequeM.append(s[i])
while dequeM.isnotempty():
    nf.write(dequeM.pop() + " ")
while dequeP.isnotempty():
    nf.write(dequeP.pop() + " ")
f.close()
nf.close()
```
