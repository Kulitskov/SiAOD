#### Бинарный поиск

```python
import random

def bn(mas_rand):
    x = int(input("Число для поиска: "))

    while len(mas_rand) > 1:
        if x > mas_rand[int(len(mas_rand)/2)]:
            mas_rand = mas_rand[int(len(mas_rand)/2):]
        elif x < mas_rand[int(len(mas_rand)/2)]:
            mas_rand = mas_rand[:int(len(mas_rand)/2)]
        else:
            return 1
    return -1

mas_rand = sorted([random.randint(1, 99) for i in range(50)])
print(mas_rand)
print(bn(mas_rand))
```

#### Бинарное дерево

```python
from random import randint

class Node:

    def __init__(self, data):

        self.left = None
        self.right = None
        self.data = data
    #создание узлы бинарнго дерева 
    def insert(self, data):
        
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
    # метод findval для сравнения значения с узлами
    def findval(self, lkpval):
        if lkpval < self.data:
            if self.left is None:
                return str(lkpval)+" Not Found"
            return self.left.findval(lkpval)
        elif lkpval > self.data:
            if self.right is None:
                return str(lkpval)+" Not Found"
            return self.right.findval(lkpval)
        else:
            return str(self.data) + " is found"

root = Node(15)

a = []
for i in range(10):
    a.append(randint(10, 25))
    root.insert(a[i])
    print(a[i], end = ' ')
print()   
print(root.findval(int(input())))
print(root.findval(int(input())))
```

#### Фибоначеев

```python
def FibonacciSearch(mas):
    x = int(input("Число для поиска: "))
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(mas)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1;
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(mas)-1))
        if (mas[i] < x):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (mas[i] > x):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else :
            return i
    if(fibM_minus_1 and index < (len(mas)-1) and mas[index+1] == x):
        return index+1;
    return -1


mas_rand = sorted([random.randint(1, 99) for i in range(50)])
print(mas_rand)
print(FibonacciSearch(mas_rand))
```

#### Интерполяционный

```python
def InterpolationSearch(mas):
    low = 0
    high = len(mas) - 1
    while low <= high and x >= mas[low] and x <= mas[high]:
        index = low + int(((float(high - low) / (mas[high] - mas[low])) * (x - mas[low])))
        if mas[index] == x:
            return index
        if mas[index] < x:
            low = index + 1
        else:
            high = index - 1
    return -1

mas_rand = sorted([random.randint(1, 99) for i in range(50)])
print(mas_rand)
print(InterpolationSearch(mas_rand))
```

#### Рехэширование с помощью псевдослучайных чисел

```python
import random

class HashClass:
    keys = []
    hashes = []
    
    def __init__(self, arr : list):
        self.appendArray(arr)
        
    def appendKey(self, item, length = None):
        if (length == None):
            length = len(self.keys)
            
        self._checkKeyExists(item)
        temp = hash(item) % length
        
        if self._checkCollisions(temp):
            temp = self._pseudoRandRehash(item, length)
            
        self.hashes.append(temp)
        self.keys.append(item)
        
    def printHashes(self):
        for i in range(len(self.keys)):
            print(self.hashes[i], self.keys[i])
        print()
        
    def appendArray(self, arr:list):
        for item in arr:
            self.appendKey(item, len(arr))
            
    def _pseudoRandRehash(self, key, length : int):
        possibleValues = [i for i in range(length * 2)]
        random.shuffle(possibleValues)
        while True:
            i = possibleValues.pop()
            temp = hash(key) % length + i
            if not self._checkCollisions(temp):
                return temp
            
    def _checkCollisions(self, hashValue):
        return hashValue in self.hashes

    def _checkKeyExists(self, key):
        if key in self.keys:
            raise Exception("Key '{}' already exists".format(key))
            
values = ['superstr', 'megastr', 'duperstr', 'supermegastr', 'supermegaduperstr', 'megaduperstr', 'test']
hashes = HashClass(arr=values)

try:
    hashes.printHashes()
    hashes.appendKey('newKey')
    hashes.appendKey('newKey2')
    hashes.appendKey('newKey3')
    hashes.appendKey('newKey4')
    hashes.printHashes()
    
    # тут будет ошибка
    hashes.appendKey('newKey')
except Exception as inst:
    print(inst.args)
```

#### Метод цепочек

```python
class ChainsMethod:
    _hashes = {}
    _values = {}
    _pointer = 0
    _N = 10
    
    def __init__(self, arr:dict = None):
        if arr != None:
            self.appendArray(arr)
    
    def append(self, key, value):
        keyHash = hash(key) % self._N
        if self._checkCollisions(keyHash):
            prevInd = self._hashes[keyHash]
            self._values[self._pointer] = [key, value, prevInd]
        else:
            self._values[self._pointer] = [key, value]
            
        self._hashes[keyHash] = self._pointer
        self._pointer += 1
            
    def appendArray(self, arr:dict):
        for key, value in arr.items():
            self.append(key, value)
            
    def find(self, key):
        keyHash = hash(key) % self._N
        if not keyHash in self._hashes:
            raise Exception('Item not found')
        index = self._hashes[keyHash]
        return self._recursiveFind(key, index)

    def _recursiveFind(self, key, index):
        if self._values[index][0] == key:
            return self._values[index][1]
        
        if len(self._values[index]) < 3:
            raise Exception('Item not found')
            
        return self._recursiveFind(key, self._values[index][2])
            
    
    def printHashes(self):
        print("index => [key, value, pervious]:")
        for key, value in self._values.items():
            print("{} => {}".format(key,value))
            
        print("\nhash => index:")
        for key, value in self._hashes.items():
            print("{} => {}".format(key,value))
        
    def _checkCollisions(self, keyHash):
        return keyHash in self._hashes
    
    
try:
    arr = {"key1" : "value1", "key2" : "value2", "kkkwer2" : "vakue24"}
    chains = ChainsMethod(arr)
    chains.printHashes()
    
    print("\n\nappend key 'newKey'")
    chains.append("newKey", "newValue")
    chains.printHashes()
    
    print("\n\nfind by key 'key2'")
    print("result: ",chains.find("key2"))
    
    # тут будет ошибка
    print("\n\nfind by key 'key'")
    print("result: ",chains.find("key"))
    
except Exception as inst:
    print(inst.args)
```
